### Analysis

- The correlation between mouse weight and the average tumor volume is 0.84, suggesting a strong positive correlation;
- Among Capomulin, Ramicane, Infubinol, and Ceftamin treatments, Ramicane has the smallest tumor volume during the final timepoint, followed by Capolmulin;
- Among Capomulin, Ramicane, Infubinol, and Ceftamin treatments, only Infubinol has one outlier which needs to be futher studied if necessary.
 

```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import scipy.stats as st
import numpy as np
from scipy.stats import linregress

# Study data files
mouse_metadata_path = "data/Mouse_metadata.csv"
study_results_path = "data/Study_results.csv"

# Read the mouse data and the study results
mouse_metadata = pd.read_csv(mouse_metadata_path)
study_results = pd.read_csv(study_results_path)

# Combine the data into a single DataFrame
df = pd.merge(study_results, mouse_metadata, how = 'outer', on = 'Mouse ID')

# Display the data table for preview
mouse_df = df.groupby('Mouse ID', as_index = False).nth(0).reset_index(drop = True)
mouse_df.head()
```

```python
# Checking the number of mice.
len(mouse_df)
```

```python
# Our data should be uniquely identified by Mouse ID and Timepoint
# Get the duplicate mice by ID number that shows up for Mouse ID and Timepoint. 
duplicated_id = df['Mouse ID'][df.duplicated(['Mouse ID','Timepoint'])].unique()
duplicated_id
```

```python
# Optional: Get all the data for the duplicate mouse ID. 
df[df['Mouse ID'] == duplicated_id[0]]
```

```python
# Create a clean DataFrame by dropping the duplicate mouse by its ID.
clean_df = df.drop(df.index[df['Mouse ID'] == duplicated_id[0]][0])
clean_df = clean_df.reset_index(drop = True)
clean_mouse_df = mouse_df.drop(mouse_df.index[mouse_df['Mouse ID'] == duplicated_id[0]][0])
clean_mouse_df = clean_mouse_df.reset_index(drop = True)
clean_mouse_df.head()
```

```python
# Checking the number of mice in the clean DataFrame.
len(clean_mouse_df)
```

## Summary Statistics

```python
# Generate a summary statistics table of mean, median, variance, standard deviation, and SEM of the tumor volume for each regimen

# Use groupby and summary statistical methods to calculate the following properties of each drug regimen: 
# mean, median, variance, standard deviation, and SEM of the tumor volume. 
# Assemble the resulting series into a single summary DataFrame.
mean_new_df = clean_df[['Drug Regimen','Tumor Volume (mm3)']].groupby('Drug Regimen').mean()
median_new_df = clean_df[['Drug Regimen','Tumor Volume (mm3)']].groupby('Drug Regimen').median()
var_new_df = clean_df[['Drug Regimen','Tumor Volume (mm3)']].groupby('Drug Regimen').var()
std_new_df = clean_df[['Drug Regimen','Tumor Volume (mm3)']].groupby('Drug Regimen').std()
sem_new_df = clean_df[['Drug Regimen','Tumor Volume (mm3)']].groupby('Drug Regimen').sem()

summary_df = pd.concat([mean_new_df, median_new_df, var_new_df, std_new_df, sem_new_df], axis = 1)
summary_df.columns = summary_df.columns = ['Mean Tumor Volume','Median Tumor Volume','Tumor Volume Variance','Tumor Volume Std. Dev.','Tumor Volume Std. Err.']
summary_df
```

```python
# A more advanced method to generate a summary statistics table of mean, median, variance, standard deviation,
# and SEM of the tumor volume for each regimen (only one method is required in the solution)

# Using the aggregation method, produce the same summary statistics in a single line
agg_df = clean_df[['Drug Regimen','Tumor Volume (mm3)']].groupby('Drug Regimen')
agg_summary_df = agg_df.aggregate(['mean','median','var','std','sem'])
agg_summary_df
```

## Quartiles, Outliers and Boxplots

```python
# Calculate the final tumor volume of each mouse across four of the treatment regimens:  
# Capomulin, Ramicane, Infubinol, and Ceftamin

# Start by getting the last (greatest) timepoint for each mouse
final_mouse_df = df.groupby('Mouse ID', as_index = False).last()
clean_final_mouse_df = final_mouse_df.drop(final_mouse_df.index[final_mouse_df['Mouse ID'] == duplicated_id[0]][0])
clean_final_mouse_df = clean_final_mouse_df.reset_index(drop = True)

# Merge this group df with the original DataFrame to get the tumor volume at the last timepoint
all_clean_mouse_df = pd.merge(clean_mouse_df,clean_final_mouse_df[['Mouse ID','Timepoint','Tumor Volume (mm3)']], how='outer', on='Mouse ID',suffixes=(' Begin',' Last'))
```

```python
final_mouse_df = df.groupby('Mouse ID', as_index = False).last()
clean_final_mouse_df = final_mouse_df.drop(final_mouse_df.index[final_mouse_df['Mouse ID'] == duplicated_id[0]][0])
clean_final_mouse_df = clean_final_mouse_df.reset_index(drop = True)
all_clean_mouse_df = pd.merge(clean_mouse_df,clean_final_mouse_df[['Mouse ID','Timepoint','Tumor Volume (mm3)']], how='outer', on='Mouse ID',suffixes=(' Begin',' Last'))
```

```python
# Put treatments into a list for for loop (and later for plot labels)
treatments = ['Capomulin', 'Ramicane', 'Infubinol', 'Ceftamin']

# Create empty list to fill with tumor vol data (for plotting)
capo_tumor_vol = []
rami_tumor_vol = []
infu_tumor_vol = []
ceft_tumor_vol = []

# Calculate the IQR and quantitatively determine if there are any potential outliers. 
capo_quartiles = all_clean_mouse_df[all_clean_mouse_df['Drug Regimen']=='Capomulin']['Tumor Volume (mm3) Last'].quantile([0.25,.5,0.75])
capo_lowerq = capo_quartiles[0.25]
capo_upperq = capo_quartiles[0.75]
capo_iqr = capo_upperq-capo_lowerq
capo_iqr_1_5 = capo_iqr * 1.5
capo_lower_bound = capo_lowerq - capo_iqr_1_5
capo_upper_bound = capo_upperq + capo_iqr_1_5

rami_quartiles = all_clean_mouse_df[all_clean_mouse_df['Drug Regimen']=='Ramicane']['Tumor Volume (mm3) Last'].quantile([0.25,.5,0.75])
rami_lowerq = rami_quartiles[0.25]
rami_upperq = rami_quartiles[0.75]
rami_iqr = rami_upperq-rami_lowerq
rami_iqr_1_5 = rami_iqr * 1.5
rami_lower_bound = rami_lowerq - rami_iqr_1_5
rami_upper_bound = rami_upperq + rami_iqr_1_5

infu_quartiles = all_clean_mouse_df[all_clean_mouse_df['Drug Regimen']=='Infubinol']['Tumor Volume (mm3) Last'].quantile([0.25,.5,0.75])
infu_lowerq = infu_quartiles[0.25]
infu_upperq = infu_quartiles[0.75]
infu_iqr = infu_upperq-infu_lowerq
infu_iqr_1_5 = infu_iqr * 1.5
infu_lower_bound = infu_lowerq - infu_iqr_1_5
infu_upper_bound = infu_upperq + infu_iqr_1_5

ceft_quartiles = all_clean_mouse_df[all_clean_mouse_df['Drug Regimen']=='Ceftamin']['Tumor Volume (mm3) Last'].quantile([0.25,.5,0.75])
ceft_lowerq = ceft_quartiles[0.25]
ceft_upperq = ceft_quartiles[0.75]
ceft_iqr = ceft_upperq-ceft_lowerq
ceft_iqr_1_5 = ceft_iqr * 1.5
ceft_lower_bound = ceft_lowerq - ceft_iqr_1_5
ceft_upper_bound = ceft_upperq + ceft_iqr_1_5

# Locate the rows which contain mice on each drug and get the tumor volumes
for treatment in treatments:
    if treatment == 'Capomulin':
        capo_tumor = all_clean_mouse_df[all_clean_mouse_df['Drug Regimen']==treatment]['Tumor Volume (mm3) Last']
        capo_tumor_vol.append(capo_tumor)
    elif treatment == 'Ramicane':
        rami_tumor = all_clean_mouse_df[all_clean_mouse_df['Drug Regimen']==treatment]['Tumor Volume (mm3) Last']
        rami_tumor_vol.append(rami_tumor)
    elif treatment == 'Infubinol':
        infu_tumor = all_clean_mouse_df[all_clean_mouse_df['Drug Regimen']==treatment]['Tumor Volume (mm3) Last']
        infu_tumor_vol.append(infu_tumor) 
    elif treatment == 'Ceftamin':
        ceft_tumor = all_clean_mouse_df[all_clean_mouse_df['Drug Regimen']==treatment]['Tumor Volume (mm3) Last']
        ceft_tumor_vol.append(ceft_tumor) 
 
    # add subset 
capo = all_clean_mouse_df[all_clean_mouse_df['Drug Regimen']=='Capomulin']['Tumor Volume (mm3) Last'].tolist()
rami = all_clean_mouse_df[all_clean_mouse_df['Drug Regimen']=='Ramicane']['Tumor Volume (mm3) Last'].tolist()
infu = all_clean_mouse_df[all_clean_mouse_df['Drug Regimen']=='Infubinol']['Tumor Volume (mm3) Last'].tolist()
ceft = all_clean_mouse_df[all_clean_mouse_df['Drug Regimen']=='Ceftamin']['Tumor Volume (mm3) Last'].tolist()

box_df = pd.DataFrame({'Capomulin':capo, 
                      'Ramicane':rami, 
                      'Infubinol':infu,
                      'Ceftamin':ceft
                      })

    
# Determine outliers using upper and lower bounds
outlier1 =[]
for i in range(25):
    if box_df['Capomulin'][i]>capo_upper_bound or box_df['Capomulin'][i]<capo_lower_bound:
        outlier1.append(box_df['Capomulin'][i])
outliers1= pd.Series(data = outlier1,name = 'Tumor Volume (mm3)', dtype=float)

outlier2 =[]   
for j in range(25):
    if box_df['Ramicane'][j]>rami_upper_bound or box_df['Ramicane'][j]<rami_lower_bound:
        outlier2.append(box_df['Ramicane'][j])
outliers2= pd.Series(data = outlier2,name = 'Tumor Volume (mm3)', dtype=float)

outlier3 =[] 
for j in range(25):
    if box_df['Infubinol'][j]>infu_upper_bound or box_df['Infubinol'][j]<infu_lower_bound:
        outlier3.append(box_df['Infubinol'][j])
outliers3= pd.Series(data = outlier3,name = 'Tumor Volume (mm3)', dtype=float)

outlier4 =[]
for j in range(25):
    if box_df['Ceftamin'][j]>ceft_upper_bound or box_df['Ceftamin'][j]<ceft_lower_bound:
        outlier4.append(box_df['Ceftamin'][j])
outliers4= pd.Series(data = outlier4,name = 'Tumor Volume (mm3)', dtype=float)

print(f"Capomulin's potential outliers: {outliers1}\nRamicane's potential outliers: {outliers2}\nInfubinol's potential outliers: {outliers3}\nCeftamin's potential outliers: {outliers4}")
```

```python
# Generate a box plot that shows the distrubution of the tumor volume for each treatment group.
capo = all_clean_mouse_df[all_clean_mouse_df['Drug Regimen']=='Capomulin']['Tumor Volume (mm3) Last'].tolist()
rami = all_clean_mouse_df[all_clean_mouse_df['Drug Regimen']=='Ramicane']['Tumor Volume (mm3) Last'].tolist()
infu = all_clean_mouse_df[all_clean_mouse_df['Drug Regimen']=='Infubinol']['Tumor Volume (mm3) Last'].tolist()
ceft = all_clean_mouse_df[all_clean_mouse_df['Drug Regimen']=='Ceftamin']['Tumor Volume (mm3) Last'].tolist()

box_df = pd.DataFrame({'Capomulin':capo, 
                      'Ramicane':rami, 
                      'Infubinol':infu,
                      'Ceftamin':ceft
                      })
circle = dict(markerfacecolor = 'red', markersize=12)
plt.boxplot(box_df,flierprops = circle )
plt.xticks(ticks = [1,2,3,4],labels = treatments)
plt.ylabel('Final Tumor Volume (mm3)')
plt.show()
```

## Line and Scatter Plots

```python
# Generate a line plot of tumor volume vs. time point for a single mouse treated with Capomulin
one_mouse = df[df['Mouse ID']=='l509']
plt.plot(one_mouse['Timepoint'],one_mouse['Tumor Volume (mm3)'])
plt.xlabel('Timepoint(days)')
plt.ylabel('Tumor Volume (mm3)')
plt.title('Capomulin treatment of mouse l509')
plt.show()
```

```python
# Generate a scatter plot of mouse weight vs. the average observed tumor volume for the entire Capomulin regimen
linear_df = clean_df[clean_df['Drug Regimen']=='Capomulin'][['Mouse ID','Weight (g)','Tumor Volume (mm3)']]
linear_df = linear_df.groupby('Mouse ID').mean()

x_values = linear_df['Weight (g)']
y_values = linear_df['Tumor Volume (mm3)']
plt.scatter(x_values,y_values)
plt.xlabel('Weight (g)')
plt.ylabel('Average Tumor Volume (mm3)')
plt.show()
```

## Correlation and Regression

```python
# Calculate the correlation coefficient and a linear regression model 
# for mouse weight and average observed tumor volume for the entire Capomulin regimen

(slope, intercept, rvalue, pvalue, stderr) = linregress(x_values, y_values)
regress_values = x_values * slope + intercept
line_eq = "y = " + str(round(slope,2)) + "x + " + str(round(intercept,2))

print(f"The correlation between mouse weight and the average tumor volume is {round(rvalue,2)}")

plt.scatter(x_values,y_values)
plt.plot(x_values,regress_values,"r-")
plt.annotate(line_eq,(5.8,0.8),fontsize=15,color="red")
plt.xlabel('Weight (g)')
plt.ylabel('Average Tumor Volume (mm3)')
plt.show()
```
